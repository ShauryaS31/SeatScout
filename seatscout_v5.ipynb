## Evaluating the performance of the model

```python
data_path = 'data.yaml'
```

### Creating a function to pretty print metrics per class

```python
def print_class_metrics(results):

    CLASSES = ['chair', 'human']

    # Print the header for per-class AP/AR table
    print("\nAP / AR per class\n" + "-"*73)
    print(f"| {'ID':<3} | {'Class':<20} | {'AP':<18} | {'AR':<18} |")
    print("-"*73)

    # Loop through each class
    for i, cls in enumerate(CLASSES):
        # Get AP and AR for the current class
        ap = results.box.maps[i]
        ar = results.box.r[i]
    
        # Print formatted row
        print(f"| {i:<3} | {cls:<20} | {ap:<18.3f} | {ar:<18.3f} |")

    # Print average AP and AR across all classes
    print("-"*73)
    print(f"| {'Avg':<24} | {results.box.map:<18.3f} | {results.box.mr:<18.3f} |")
    print("-"*73)
```

### Training set performance

```python
train_results = model.val(data=data_path, split='train')
print_class_metrics(train_results)
```

### Validation set performance

```python
val_results = model.val(data=data_path, split='val')
print_class_metrics(val_results)
```

### Test set performance

```python
test_results = model.val(data=data_path, split='test')
print_class_metrics(test_results)
```

### Inference example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2

img_path = 'data/dataset_v2/images/test/59.png' #'sample_image.png'

def inference(img_path):
# Predict
    results = model.predict(source=img_path, conf=0.6)
    result = results[0]

    # Pull clean data
    img_bgr = result.orig_img
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    boxes = result.boxes.xyxy.cpu()
    scores = result.boxes.conf.cpu()
    labels = result.boxes.cls.cpu().int()

    class_colors = {
        "human": "red",
        "chair": "green",
        "table": "#33C1FF",
    }

    plt.figure(figsize=(12, 8))
    plt.imshow(img_rgb)
    ax = plt.gca()

    for (x1, y1, x2, y2), score, label in zip(boxes, scores, labels):
        if score > 0.6:
            class_name = model.names[label.item()]
            color = class_colors.get(class_name, 'yellow')  # default color
            ax.add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1,
                                    edgecolor=color, facecolor='none', linewidth=2))
            ax.text(x1, y1-5, f'{class_name}: {score:.2f}',
                    color='white', fontsize=10,
                    bbox=dict(facecolor=color, alpha=0.5, edgecolor='none'))

    plt.axis('off')
    plt.title(f'Predictions for {img_path.split("/")[-1]}')
    plt.show()

inference(img_path)
```
